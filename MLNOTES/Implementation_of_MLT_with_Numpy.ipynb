### K-Means Clustering Algorithm

K-means clustering is an unsupervised machine learning algorithm that partitions n observations into k clusters.

**Algorithm Steps:**
1. Initialize k centroids randomly
2. Assignment: Assign each point to nearest centroid
3. Update: Recalculate centroids as mean of assigned points
4. Repeat steps 2-3 until convergence

**Objective:** Minimize within-cluster sum of squares (WCSS):

$$ J = \sum_{k=1}^{K} \sum_{x_i \in C_k} ||x_i - \mu_k||^2 $$

where $\mu_k$ is the centroid of cluster $C_k$

#### K-Means Implementation from Scratch (NumPy)

```
# K-Means Clustering from Scratch using NumPy

def euclidean_distance(x1, x2):
    """Calculate Euclidean distance between two points"""
    return np.sqrt(np.sum((x1 - x2)**2))

def initialize_centroids(X, K):
    """Randomly initialize K centroids from the dataset"""
    m, n = X.shape
    centroids = np.zeros((K, n))
    rand_idx = np.random.choice(m, K, replace=False)
    centroids = X[rand_idx, :]
    return centroids

def closest_centroid(x, centroids, K):
    """Find the closest centroid for a given data point"""
    distances = np.zeros(K)
    for k in range(K):
        distances[k] = euclidean_distance(x, centroids[k])
    return np.argmin(distances)

def create_clusters(X, centroids, K):
    """Assign each data point to the closest centroid"""
    m, _ = X.shape
    cluster_idx = np.zeros(m, dtype=int)
    for i in range(m):
        cluster_idx[i] = closest_centroid(X[i], centroids, K)
    return cluster_idx

def compute_centroids(X, cluster_idx, K):
    """Compute new centroids as mean of points in each cluster"""
    _, n = X.shape
    centroids = np.zeros((K, n))
    for k in range(K):
        points = X[cluster_idx == k]
        if len(points) > 0:
            centroids[k] = np.mean(points, axis=0)
    return centroids

def kmeans(X, K, max_iterations=100):
    """K-Means clustering algorithm"""
    # Initialize centroids
    centroids = initialize_centroids(X, K)
    
    for iteration in range(max_iterations):
        # Assign points to clusters
        cluster_idx = create_clusters(X, centroids, K)
        
        # Store old centroids
        old_centroids = centroids.copy()
        
        # Compute new centroids
        centroids = compute_centroids(X, cluster_idx, K)
        
        # Check for convergence
        if np.allclose(old_centroids, centroids):
            print(f"Converged at iteration {iteration+1}")
            break
    
    return cluster_idx, centroids

print("K-Means functions defined successfully!")
```

#### Generate Sample Data for K-Means

```
# Generate sample data with 3 clusters
np.random.seed(42)

# Cluster 1: centered around (2, 2)
cluster1 = np.random.randn(100, 2) + np.array([2, 2])

# Cluster 2: centered around (8, 3)
cluster2 = np.random.randn(100, 2) + np.array([8, 3])

# Cluster 3: centered around (5, 8)
cluster3 = np.random.randn(100, 2) + np.array([5, 8])

# Combine all clusters
X = np.vstack([cluster1, cluster2, cluster3])

print(f"Dataset shape: {X.shape}")
print(f"First 5 data points:\n{X[:5]}")
```

#### Run K-Means Clustering (NumPy Implementation)

```
# Run K-Means with K=3 clusters
K = 3
cluster_labels, final_centroids = kmeans(X, K, max_iterations=100)

print(f"\nFinal centroids:\n{final_centroids}")
print(f"\nCluster assignments (first 10): {cluster_labels[:10]}")
print(f"\nCluster distribution:")
for k in range(K):
    count = np.sum(cluster_labels == k)
    print(f"  Cluster {k}: {count} points")
```

#### Visualize K-Means Results

```
# Visualize clustering results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green']

for k in range(K):
    cluster_points = X[cluster_labels == k]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                c=colors[k], label=f'Cluster {k}', alpha=0.5)

# Plot centroids
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], 
            c='black', marker='X', s=200, label='Centroids')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering Results (K=3)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

#### K-Means Using Scikit-Learn

```
# K-Means using scikit-learn
from sklearn.cluster import KMeans

# Create and fit K-Means model
kmeans_sklearn = KMeans(n_clusters=3, init='k-means++', 
                         max_iter=100, n_init=10, random_state=42)
sklearn_labels = kmeans_sklearn.fit_predict(X)

print(f"Scikit-learn centroids:\n{kmeans_sklearn.cluster_centers_}")
print(f"\nInertia (WCSS): {kmeans_sklearn.inertia_:.2f}")
print(f"Number of iterations: {kmeans_sklearn.n_iter_}")
```

#### Elbow Method to Find Optimal K

```
# Elbow method to determine optimal number of clusters
inertias = []
K_range = range(1, 11)

for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, init='k-means++', 
                         max_iter=100, n_init=10, random_state=42)
    kmeans_temp.fit(X)
    inertias.append(kmeans_temp.inertia_)

# Plot elbow curve
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Inertia (WCSS)', fontsize=12)
plt.title('Elbow Method For Optimal K', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(K_range)
plt.show()

print("The 'elbow' point suggests the optimal number of clusters")
```

#### K-Means on Real Data Example

```
# Example: Clustering the Iris dataset
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load Iris dataset
iris = load_iris()
X_iris = iris.data
y_true = iris.target

# Standardize features
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)

# Apply K-Means
kmeans_iris = KMeans(n_clusters=3, init='k-means++', 
                     max_iter=300, n_init=10, random_state=42)
y_pred = kmeans_iris.fit_predict(X_iris_scaled)

print(f"Iris dataset shape: {X_iris.shape}")
print(f"\nCluster centers shape: {kmeans_iris.cluster_centers_.shape}")
print(f"Inertia: {kmeans_iris.inertia_:.2f}")
print(f"\nPredicted cluster distribution:")
for k in range(3):
    count = np.sum(y_pred == k)
    print(f"  Cluster {k}: {count} samples")
```

```
# Visualize Iris clustering (using first 2 features)
plt.figure(figsize=(12, 5))

# Plot 1: True labels
plt.subplot(1, 2, 1)
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=y_true, cmap='viridis', alpha=0.6)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('True Iris Species')
plt.colorbar(label='Species')

# Plot 2: K-Means predictions
plt.subplot(1, 2, 2)
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=y_pred, cmap='viridis', alpha=0.6)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-Means Clustering')
plt.colorbar(label='Cluster')

plt.tight_layout()
plt.show()
```
